```python
import math
import numpy as np
import matplotlib.pyplot as plt
import jax
import jax.numpy as jnp

import sys
import os
notebook_dir = os.path.dirname(os.path.realpath("__file__"))
sys.path.append(os.path.join(notebook_dir, '..', 'src'))

from mip_ml.modeling import Exp, Log, Logistic, Hill, make_pwl_pts
```

## Log

```python
LB, UB = 1, 100
f = Log(LB,UB)

xs = np.linspace(LB,UB)
ys = f(xs)
x_apx = make_pwl_pts(f, tol=1e-1)
y_apx = f(x_apx)

plt.plot(xs, ys)
plt.plot(x_apx, y_apx, marker = 'o')
```

## Exp

```python
LB, UB = -10, 5
f = Exp(LB, UB)

xs = np.linspace(LB,UB)
ys = f(xs)
x_apx = make_pwl_pts(f, tol=1)
y_apx = f(x_apx)

plt.plot(xs, ys)
plt.plot(x_apx, y_apx, marker = 'o')
```

## Logistic

```python
LB, UB = -10, 10
f = Logistic(LB, UB)

xs = np.linspace(LB,UB)
ys = f(xs)
x_apx = make_pwl_pts(f, tol=1e-2)
y_apx = f(x_apx)

plt.plot(xs, ys)
plt.plot(x_apx, y_apx, marker = 'o')
```

## Hill

```python
LB, UB = 0, 100
f = Hill(LB, UB, params={'a':2,'b':0.1,'c':10})

xs = np.linspace(LB,UB)
ys = f(xs)
x_apx = make_pwl_pts(f, tol=1e-2)
y_apx = f(x_apx)

plt.plot(xs, ys)
plt.plot(x_apx, y_apx, marker = 'o')
```
